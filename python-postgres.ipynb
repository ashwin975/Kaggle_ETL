```python
import psycopg2
import pandas as pd
from psycopg2 import extras

databasename='postgres'
username='postgres'
pwd='admin'
hostname='localhost'
port_id='5432'



try:
    conn = psycopg2.connect(
        dbname=databasename,
        user=username,
        password=pwd,
        host=hostname,
        port=port_id
    )
    cur = conn.cursor()
    

    cur.execute("DROP TABLE IF EXISTS avocsss")

    create_script = """CREATE TABLE avocsss (
                        Date DATE,
                        AveragePrice FLOAT,
                        TotalVolume FLOAT,
                        plu_4046 double precision,
                        plu_4225 double precision,
                        plu_4770 double precision,
                        TotalBags FLOAT,
                        SmallBags FLOAT,
                        LargeBags FLOAT,
                        XLargeBags FLOAT,
                        type VARCHAR(50),
                        year INT,
                        region VARCHAR(50),
                        sequence INT PRIMARY KEY
                    );"""

    cur.execute(create_script)
    conn.commit()
    
    psycopg2.extras.execute_values(
        cur,
        """
        INSERT INTO avocsss (
            Date, AveragePrice, TotalVolume, plu_4046, plu_4225, plu_4770, TotalBags, SmallBags, LargeBags, XLargeBags, type, year, region, sequence
        ) VALUES %s
        """,
        df.values.tolist()
    )
    conn.commit()
    
except (Exception, psycopg2.DatabaseError) as error:
    print("Error occurred:", error)

finally:
    if conn is not None:
        if cur is not None:
            cur.close()
        conn.close()
```
